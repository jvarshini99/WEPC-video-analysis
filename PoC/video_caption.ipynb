```python
import re
import os
import glob
import fnmatch
import pathlib
import ffmpeg
```

Helper code that returns directory of all filepaths to the videos

```python
path = "../teach-forward-videos"
pattern = "*.mp4"

file_list = []

for dirpath, dirnames, filenames in os.walk(path):
    if not filenames:
        continue

    files = fnmatch.filter(filenames, pattern)
    if files:
        for file in files:
            file_with_path = '{}/{}'.format(dirpath, file)
            file_list.append(file_with_path)
print(file_list)
```

Uses ffmpeg to cut a one-minute clip of each video for downstream captioning

```python
k = 0
for filename in file_list:
    filename_rel = filename[2:]
    #print(filename)
    !ffmpeg -i /projectnb/sparkgrp/ml-wheelock-video-grp{filename_rel} -ss 00:00:00 -to 00:01:00 -c:v copy -c:a copy /projectnb/sparkgrp/ml-wheelock-video-grp/PDVC/video_clips/{k}.mp4
    k += 1
```
